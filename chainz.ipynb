```
import pandas as pd
```

```
#create empty df
df = pd.DataFrame()

#create a column
df['name'] = ['Damian', 'Sarah', 'Pe6o']
df['employed'] = ['Yes', 'No', 'Yes']
df['age'] = [32,29,19]

#view df
df
```

```
#create func
def mean_age_by_group(dataframe, col):
  #groups the data by a col and return the mean age per group
  yield dataframe.groupby(col).mean()
```

```
def uppercase_column_name(dataframe):
  dataframe.columns = dataframe.columns.str.upper()
  return dataframe
```

```
(df.pipe(mean_age_by_group,col='employed')
    .pipe(uppercase_column_name)
)
```
