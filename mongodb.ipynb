```
def runtime(a, b):
    """Expect two object of type datetime.datetime.time();
        return last known time run time adfdsafd"""

    t1 = timedelta(a.hour, a.second, a.microsecond)
    t2 = timedelta(b.hour, b.second, b.microsecond)
    c = t2 - t1
    return c # .total_seconds()
```
